```python
#Basically Importing the needed libraries 
#sklearn.datasets is a group of sample datasets that come with the library and are just useful to lean and get familiar
#With the concepts

import numpy as np 
import pandas as pd 
import seaborn as sns 
import sklearn.datasets
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor 
```

```python
#This the one of the sample datasets being loaded into a variable so that we can learn 
# the load_boston is a command for loading the boston housing data command 

house_price_data = sklearn.datasets.load_boston()
```

```python
# Printing the boston housing data contents 

print(house_price_data)
```

```python
# Our boston data cant be used the way it is , so we convert into a pandas dataframe 
# Wrote DataFrame as Dataframe and got an error, watch oput for that 
# If our data has column titles or features , we should include the columns = data.feature name part
# It is wise to seperate what our prediction feature since thats the feature we are predicting , in this case , Price



house_data = pd.DataFrame(house_price_data.data, columns = house_price_data.feature_names)
house_data['price'] = house_price_data.target
house_data.head()
```

```python
# checking the number of rows and columns in the data
house_data.shape
```

```python
#check for missing values 
house_data.isnull().sum()
```

```python
# obtain the stats of the data 
house_data.describe()
```

```python
#Find correlation between the the features 
correlation = house_data.corr()

#construct a heatmap to understand the correlation 

plt.figure(figsize = (12,12 ))
sns.heatmap(correlation, cbar=True, fmt='.2f', annot=True, annot_kws = {'size':10}, cmap='Blues' )
```

```python
#Time to split the data into labels 
# First we assign the features we are training with into two(x,y)
# This means , regardless of the number of features, all of them will be placed in the X-variable
# and the feature we are trying to predict (price) will be the  y-variable 
# if we are dropping a column, axis = 1, row, axis =  0
X = house_data.drop (['price'], axis = 1)
Y = house_data['price']

print(X,Y )
```

```python
#splitting the data to test and train 

x_train, y_train , x_test, y_test = train_test_split(X,Y, test_size= 0.2, random_state = 2)
```
